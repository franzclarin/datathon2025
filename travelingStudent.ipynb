```python
from scipy.spatial.distance import pdist, squareform

def dist_haversine(coords1, coords2):
    lat1, lon1 = coords1
    lat2, lon2 = coords2

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    R = 6371  
    return R * c

coords = np.array(data[["Latitude", "Longitude"]])
coords_rad = np.radians(coords)
dist_matrix = squareform(pdist(coords_rad, metric=dist_haversine)) 
dist_matrix 
```

```python
from scipy.spatial.distance import squareform, pdist
import numpy as np

# Make sure RecordID is integer (or indexable)
data['RecordID'] = data['RecordID'].astype(int)

# Get coordinates and compute full distance matrix
coords = np.radians(data[["Latitude", "Longitude"]])
dist_matrix = pd.DataFrame(
    squareform(pdist(coords, metric=lambda u, v: dist_haversine(u, v))),
    index=data['RecordID'].values,
    columns=data['RecordID'].values
)

# Get unique streets
streets = pd.unique(data['Street'])

# Store optimal orderings
optimal_paths = {}

# Greedy TSP function
def greedy_tsp(dist_submatrix):
    nodes = list(dist_submatrix.index)
    path = [nodes[0]]
    remaining = set(nodes[1:])
    while remaining:
        last = path[-1]
        next_node = min(remaining, key=lambda x: dist_submatrix.loc[last, x])
        path.append(next_node)
        remaining.remove(next_node)
    return path

# Solve TSP for each street
for street in streets:
    subset = data[data['Street'] == street]
    ids = subset['RecordID'].tolist()
    
    if len(ids) <= 1:
        optimal_paths[street] = ids
        continue

    dist_submatrix = dist_matrix.loc[ids, ids]
    path = greedy_tsp(dist_submatrix)
    optimal_paths[street] = path

# Now optimal_paths is a dict: { street_name: [RecordID1, RecordID2, ...] }
optimal_paths
```
